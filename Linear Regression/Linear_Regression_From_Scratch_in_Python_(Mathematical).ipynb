```python
def gradient_desecnt(m_now, b_now, points, L):
  m_gradient = 0
  b_gradient = 0

  n = len(points)

  for i in range(n):
    x = points.iloc[i].StudyTime
    y = points.iloc[i].Score

    m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
    b_gradient += -(2/n) * (y - (m_now * x + b_now))

  m = m_now - m_gradient * L
  b = b_now - b_gradient * L

  return m, b
```

```python
m = 0
b = 0
L = 0.0001
epochs = 300

for i in range(epochs):
  if i % 50 == 0:
    print(f"Epoch: {i}")
  m, b = gradient_desecnt(m, b, df, L)

print(m,b)

plt.scatter(df.StudyTime, df.Score, color='Black')
plt.plot(list(range(30, 65)), [m * x + b for x in range(30, 65)], color='red')
plt.show()
```

| Concept           | Like...                                                     |
| ----------------- | ----------------------------------------------------------- |
| Linear Regression | Drawing a line to guess answers from past data              |
| Cost Function     | Measuring how wrong your guesses are                        |
| Gradient Descent  | A robot adjusting the line step-by-step to fix your guesses |


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load dataset
df = pd.read_csv('data.csv')

# 2. Extract features (x) and target (y)
x = df['StudyTime'].values
y = df['Score'].values

# 3. Initialize parameters
m, b = 0, 0
lr = 0.0001
epochs = 300

# 4. Gradient Descent Loop
for _ in range(epochs):
    y_pred = m * x + b
    error = y - y_pred

    m_grad = -(2 / len(x)) * np.sum(x * error)
    b_grad = -(2 / len(x)) * np.sum(error)

    m -= lr * m_grad
    b -= lr * b_grad

# 5. Plot the final regression line
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, m * x + b, color='green', label=f'Best Fit Line: y = {m:.2f}x + {b:.2f}')
plt.title("Linear Regression with Gradient Descent")
plt.xlabel("Study Time (hours)")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()
```
